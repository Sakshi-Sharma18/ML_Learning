```python
sns.relplot(x="city_mpg", y="highway_mpg", hue="body_style", data = df)
```

```python
sns.catplot(x="body_style", y="symboling", jitter = False,  data=df)
```

```python
sns.catplot(x="body_style", y="symboling", kind="swarm" ,s=1,  data=df)
```
